# Feature 4: regates completos p90

```python
import pandas as pd
import numpy as np
```

```python
laLiga = pd.read_csv(r'C:\Users\DELUX\Desktop\datos_OPTA\laLiga.csv')
```

```python
jugadas = laLiga[['categoria_evento','fecha','id_evento','id_periodo','minuto','segundo','equipo','calificador','valor','id_jugador','jugador','x','y','outcome']].copy()
jugadas['valor'].fillna(True, inplace=True)
```

```python
jugadas.tail()
```

```python
jugadas.drop(jugadas.index[[177812]], inplace=True)
```

```python
jugadas = jugadas.set_index(['categoria_evento','fecha','id_evento','id_periodo','minuto','segundo','equipo','id_jugador','jugador','x','y','outcome','calificador'])['valor'].unstack()
jugadas = jugadas.reset_index().rename_axis(None).rename_axis(None, axis=1)
jugadas.fillna(False, inplace=True)
```

```python
jugadas.head()
```

```python
jugadas = jugadas[jugadas['jugador']!=False]
```

```python
jugadas = jugadas[['fecha','equipo','id_jugador','jugador','categoria_evento','outcome']]
```

```python
jugadas.head()
```

```python
jugadas = jugadas[(jugadas['categoria_evento']=='drible')&(jugadas['outcome']==1)]
```

```python
jugadas.head()
```

```python
mtr4 = jugadas.groupby(['equipo','id_jugador','jugador']).count().reset_index()
mtr4 = mtr4.drop(['categoria_evento','fecha'],1)
mtr4.rename(columns={'outcome':'mtr4'},inplace=True)
```

```python
mtr4.sort_values('mtr4',ascending=False).head()
```

```python
dataAggJug = pd.read_csv(r'C:\Users\DELUX\Desktop\consultoria_futbol\comparativoJugadores\dataAggJug.csv')
```

```python
dataAggJug.head()
```

```python
minJug = dataAggJug[dataAggJug['estadistica']=='mins_played']
```

```python
minJug.head()
```

```python
minJug = minJug.drop('estadistica',1).rename(columns={'num_estadistica':'minJug'})
```

```python
minJug.head()
```

```python
minJug = minJug.groupby(['id_jugador','jugador']).sum().reset_index().drop('fecha',1)
```

```python
minJug.head()
```

```python
mtr4x90 = minJug.merge(mtr4,on=['id_jugador','jugador'],how='left') #Hacemos left join para tener tabla con minutos jugados por jugador por fecha
mtr4x90['mtr4x90'] = mtr4x90['mtr4']/mtr4x90['minJug']*90 #Creamos mtr2x90
mtr4x90 = mtr4x90.drop(['mtr4','minJug'],1)
```

```python
mtr4x90.head()
```

```python
mtr4x90.isnull().sum()
```

```python
mtr4x90['mtr4x90'] = mtr4x90['mtr4x90'].replace(np.nan, 0)
```

```python
mtr4x90.isnull().sum()
```

```python
mtr4x90.head()
```

```python
feat4 = mtr4x90.copy()
feat4 = feat4.rename(columns={'mtr4x90':'feat4'})
```
