# Feature 6: balones perdidos por mal control p90 (turnovers pt2)

```python
import numpy as np
import pandas as pd
```

```python
laLiga = pd.read_csv(r'C:\Users\DELUX\Desktop\datos_OPTA\laLiga.csv')
```

```python
jugadas = laLiga[['id_jugador','categoria_evento','fecha','id_evento','id_periodo','minuto','segundo','equipo','calificador','valor','jugador','x','y','outcome']].copy()
jugadas['valor'].fillna(True, inplace=True)
```

```python
jugadas.drop(jugadas.index[[177812]], inplace=True)
```

```python
jugadas = jugadas.set_index(['id_jugador','categoria_evento','fecha','id_evento','id_periodo','minuto','segundo','equipo','jugador','x','y','outcome','calificador'])['valor'].unstack()
jugadas = jugadas.reset_index().rename_axis(None).rename_axis(None, axis=1)
jugadas.fillna(False, inplace=True)
```

```python
jugadas = jugadas[jugadas['jugador']!=False]
```

```python
jugadas = jugadas[['fecha','equipo','id_jugador','jugador','categoria_evento','outcome']]
```

```python
jugadas = jugadas[(jugadas['categoria_evento']=='mala recepción')&(jugadas['outcome']==0)]
```

```python
jugadas.head()
```

```python
jugadas[jugadas['jugador']=='Lionel Messi'].head()
```

```python
mtr6 = jugadas.groupby(['equipo','id_jugador','jugador']).count().drop(['fecha','categoria_evento'],1).rename(columns={'outcome':'mtr6'}).reset_index()
```

```python
mtr6.head()
```

```python
mtr6[mtr6['jugador']=='Lionel Messi']
```

```python
dataAggJug = pd.read_csv(r'C:\Users\DELUX\Desktop\consultoria_futbol\comparativoJugadores\dataAggJug.csv')
```

```python
dataAggJug.head()
```

```python
minJug = dataAggJug[dataAggJug['estadistica']=='mins_played']
```

```python
minJug.head()
```

```python
minJug = minJug.drop('estadistica',1).rename(columns={'num_estadistica':'minJug'})
```

```python
minJug.head()
```

```python
minJug = minJug.groupby(['id_jugador','jugador']).sum().reset_index().drop('fecha',1)
```

```python
minJug.head()
```

```python
mtr6x90 = minJug.merge(mtr6,on=['id_jugador','jugador'],how='left') #Hacemos left join para tener tabla con minutos jugados por jugador por fecha
mtr6x90['mtr6x90'] = mtr6x90['mtr6']/mtr6x90['minJug']*90 #Creamos mtr2x90
mtr6x90 = mtr6x90.drop(['mtr6','minJug'],1)
```

```python
mtr6x90.head()
```

```python
mtr6x90[mtr6x90['jugador']=='Lionel Messi']
```

```python
mtr6x90.isnull().sum()
```

```python
mtr6x90['mtr6x90'] = mtr6x90['mtr6x90'].replace(np.nan, 0)
```

```python
mtr6x90.isnull().sum()
```

```python
mtr6x90.tail()
```

```python
feat6 = mtr6x90.copy()
feat6 = feat6.rename(columns={'mtr6x90':'feat6'})
```
