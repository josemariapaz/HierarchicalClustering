# Feature 3: % pases completos

```python
import pandas as pd
import numpy as np
```

```python
pases = pd.read_csv(r'C:\Users\DELUX\Desktop\consultoria_futbol\prototipo\versiones3\pases.csv',low_memory=False)
```

```python
pases = pases[['fecha','equipo','id_jugador','jugador','outcome']] #descartamos las columnas que no nos interesan
```

```python
pases.head()
```

```python
pases = pases.groupby(['equipo','id_jugador','jugador','outcome']).count().reset_index()
pases.rename(columns={'fecha':'cant_pases'},inplace=True)
```

```python
pases.head()
```

```python
pases[pases['jugador']=='Lionel Messi']
```

```python
pases = pases.pivot_table('cant_pases',['equipo','id_jugador','jugador'],'outcome').reset_index()
```

```python
pases.head()
```

```python
pases.isnull().sum()
```

```python
pases[pases['jugador']=='Cristo González']
```

```python
pases[0] = pases[0].replace(np.nan, 0)
pases[1] = pases[1].replace(np.nan, 0)
```

```python
pases.isnull().sum()
```

```python
pases[pases['jugador']=='Cristo González']
```

```python
pases['mtr3'] = pases[1] / (pases[0]+pases[1])
```

```python
pases.head()
```

```python
pases = pases.drop([0,1],1).rename(columns={'mtr3':'feat3'})
```

```python
pases.head()
```

```python
pases[pases['equipo']=='Barcelona']
```
