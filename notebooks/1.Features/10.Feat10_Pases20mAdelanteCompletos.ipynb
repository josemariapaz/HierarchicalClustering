# Feature 10: pases largos (<20m) hacía adelante completos p90

```python
# librearías
import pandas as pd
import numpy as np
```

```python
pases = pd.read_csv(r'C:\Users\DELUX\Desktop\consultoria_futbol\prototipo\versiones3\pases.csv',low_memory=False)
```

```python
pases = pases.drop(pases[(pases['outcome']==0)| #descartamos los pases incompletos
                        (pases['adelante']==False)].index)
pases = pases[['fecha','equipo','id_jugador','jugador','adelante','distancia de balón']] #descartamos las columnas que no nos interesan
```

```python
pases.columns
```

```python
pases.head()
```

```python
pases['<20m'] = np.where(20<pases['distancia de balón'],True,False)
pases = pases.drop(pases[(pases['<20m']==False)].index)
```

```python
pases.head()
```

```python
mtr10 = pases.groupby(['fecha','equipo','id_jugador','jugador']).count().reset_index().drop(['adelante','<20m'],1).rename(columns={'distancia de balón':'mtr10'})
```

```python
mtr10.head()
```

```python
mtr10[mtr10['jugador']=='Lionel Messi'].head()
```

```python
mtr10 = mtr10.groupby(['equipo','id_jugador','jugador']).sum().reset_index().drop('fecha',1)
```

```python
mtr10.head()
```

```python
mtr10[mtr10['jugador']=='Lionel Messi']
```

```python
mtr10.sort_values('mtr10',ascending=False).head(10)
```

```python
dataAggJug = pd.read_csv(r'C:\Users\DELUX\Desktop\consultoria_futbol\comparativoJugadores\dataAggJug.csv')
```

```python
dataAggJug.head()
```

```python
minJug = dataAggJug[dataAggJug['estadistica']=='mins_played']
```

```python
minJug = minJug.drop('estadistica',1).rename(columns={'num_estadistica':'minJug'})
```

```python
minJug.head()
```

```python
minJug[minJug['jugador']=='Ricard Puig']
```

```python
minJug = minJug.groupby(['id_jugador','jugador']).sum().reset_index().drop('fecha',1)
```

```python
minJug.head()
```

```python
minJug[minJug['jugador']=='Iván López Álvarez']
```

```python
mtr10x90 = minJug.merge(mtr10,on=['id_jugador','jugador'],how='left') #Hacemos left join para tener tabla con minutos jugados por jugador por fecha
mtr10x90['mtr10x90'] = mtr10x90['mtr10']/mtr10x90['minJug']*90 #Creamos mtr10x90
mtr10x90 = mtr10x90.drop(['mtr10','minJug'],1)
```

```python
mtr10x90.head(10)
```

```python
mtr10x90.isnull().sum()
```

```python
mtr10x90['mtr10x90'] = mtr10x90['mtr10x90'].replace(np.nan, 0)
```

```python
mtr10x90.head()
```

```python
mtr10x90.isnull().sum()
```

```python
mtr10x90[mtr10x90['jugador']=='Raúl García']
```

```python
feat10 = mtr10x90.copy()
feat10 = feat10.rename(columns={'mtr10x90':'feat10'})
```
