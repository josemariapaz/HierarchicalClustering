# Feature 8: pases completos hacía adelante en el tercio final p90

```python
# librearías
import pandas as pd
import numpy as np
```

```python
pases = pd.read_csv(r'C:\Users\DELUX\Desktop\consultoria_futbol\prototipo\versiones3\pases.csv',low_memory=False)
```

```python
pases = pases.drop(pases[(pases['outcome']==0)| #descartamos los pases incompletos
                        (pases['coord_x_0']==1)|(pases['coord_x_0']==2)|(pases['coord_x_0']==3)|(pases['coord_x_0']==4)| #descartamos los pases originados fuera del tercio final
                        (pases['adelante']==False)].index)
pases = pases[['fecha','equipo','id_jugador','jugador','coord_x_0','adelante']] #descartamos las columnas que no nos interesan
```

```python
pases.columns
```

```python
pases.head()
```

```python
mtr8 = pases.groupby(['fecha','equipo','id_jugador','jugador']).count().reset_index().drop('coord_x_0',1).rename(columns={'adelante':'mtr8'})
#mtr8.rename(columns={'coord_x_1':'mtr8'},inplace=True)
```

```python
mtr8.head()
```

```python
mtr8[mtr8['jugador']=='Lionel Messi'].head()
```

```python
mtr8 = mtr8.groupby(['equipo','id_jugador','jugador']).sum().reset_index().drop('fecha',1)
```

```python
mtr8.head()
```

```python
mtr8[mtr8['jugador']=='Lionel Messi']
```

```python
mtr8.sort_values('mtr8',ascending=False).head(10)
```

```python
dataAggJug = pd.read_csv(r'C:\Users\DELUX\Desktop\consultoria_futbol\comparativoJugadores\dataAggJug.csv')
```

```python
dataAggJug.head()
```

```python
minJug = dataAggJug[dataAggJug['estadistica']=='mins_played']
```

```python
minJug = minJug.drop('estadistica',1).rename(columns={'num_estadistica':'minJug'})
```

```python
minJug.head()
```

```python
minJug[minJug['jugador']=='Carles Aleñá']
```

```python
minJug = minJug.groupby(['id_jugador','jugador']).sum().reset_index().drop('fecha',1)
```

```python
minJug.head()
```

```python
minJug[minJug['jugador']=='Iván López Álvarez']
```

```python
mtr8x90 = minJug.merge(mtr8,on=['id_jugador','jugador'],how='left') #Hacemos left join para tener tabla con minutos jugados por jugador por fecha
mtr8x90['mtr8x90'] = mtr8x90['mtr8']/mtr8x90['minJug']*90 #Creamos mtr8x90
mtr8x90 = mtr8x90.drop(['mtr8','minJug'],1)
```

```python
mtr8x90.head(10)
```

```python
mtr8x90.isnull().sum()
```

```python
mtr8x90['mtr8x90'] = mtr8x90['mtr8x90'].replace(np.nan, 0)
```

```python
mtr8x90.head()
```

```python
mtr8x90.isnull().sum()
```

```python
feat8 = mtr8x90.copy()
feat8 = feat8.rename(columns={'mtr8x90':'feat8'})
```
