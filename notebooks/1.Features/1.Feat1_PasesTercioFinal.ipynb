# Feature 1: pases al tercio final p90

```python
# librearías
import pandas as pd
import numpy as np
```

```python
pases = pd.read_csv(r'C:\Users\DELUX\Desktop\consultoria_futbol\prototipo\versiones3\pases.csv')
```

```python
pases = pases.drop(pases[(pases['outcome']==0)| #descartamos los pases incompletos
                         (pases['coord_x_0']==5)|(pases['coord_x_0']==6)| #descartamos los pases realizados desde el tercio final
                                  (pases['coord_x_1']==1)|(pases['coord_x_1']==2)|(pases['coord_x_1']==3)|(pases['coord_x_1']==4)].index) #descartamos los pases con destinos diferentes del tercio final
pases = pases[['fecha','equipo','id_jugador','jugador','coord_x_0','coord_x_1']] #descartamos las columnas que no nos interesan
```

```python
pases.columns
```

```python
mtr1 = pases.groupby(['fecha','equipo','id_jugador','jugador']).count().reset_index().drop('coord_x_0',1)
mtr1.rename(columns={'coord_x_1':'mtr1'},inplace=True)
```

```python
mtr1.head()
```

```python
mtr1[mtr1['jugador']=='Mubarak Wakaso']
```

```python
mtr1 = mtr1.groupby(['equipo','id_jugador','jugador']).sum().reset_index().drop('fecha',1)
```

```python
mtr1.head()
```

```python
mtr1[mtr1['jugador']=='Mubarak Wakaso']
```

```python
dataAggJug = pd.read_csv(r'C:\Users\DELUX\Desktop\consultoria_futbol\comparativoJugadores\dataAggJug.csv')
```

```python
dataAggJug.head()
```

```python
minJug = dataAggJug[dataAggJug['estadistica']=='mins_played']
```

```python
minJug = minJug.drop('estadistica',1).rename(columns={'num_estadistica':'minJug'})
```

```python
minJug.head()
```

```python
minJug[minJug['jugador']=='Carles Aleñá']
```

```python
minJug = minJug.groupby(['id_jugador','jugador']).sum().reset_index().drop('fecha',1)
```

```python
minJug.head()
```

```python
minJug[minJug['jugador']=='Carles Aleñá']
```

```python
mtr1x90 = minJug.merge(mtr1,on=['id_jugador','jugador'],how='left') #Hacemos left join para tener tabla con minutos jugados por jugador por fecha
mtr1x90['mtr1x90'] = mtr1x90['mtr1']/mtr1x90['minJug']*90 #Creamos mtr1x90
mtr1x90 = mtr1x90.drop(['mtr1','minJug'],1)
```

```python
mtr1x90['mtr1x90'] = mtr1x90['mtr1x90'].replace(np.nan, 0)
```
