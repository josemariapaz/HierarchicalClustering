```python
# librearías
import pandas as pd
import numpy as np
```

```python
laLiga = pd.read_csv(r'C:\Users\DELUX\Desktop\datos_OPTA\laLiga.csv')
```

```python
toques = laLiga[['categoria_evento','fecha','id_evento','id_periodo','minuto','segundo','equipo','calificador','valor','id_jugador','jugador','x','y','outcome']].copy()
toques['valor'].fillna(True, inplace=True)
```

```python
toques.drop(toques.index[[177812]], inplace=True)
```

```python
toques = toques.set_index(['categoria_evento','fecha','id_evento','id_periodo','minuto','segundo','equipo','id_jugador','jugador','x','y','outcome','calificador'])['valor'].unstack()
toques = toques.reset_index().rename_axis(None).rename_axis(None, axis=1)
toques.fillna(False, inplace=True)
```

```python
toques = toques[toques['jugador']!=False]
```

```python
toques = toques[['fecha','equipo','id_jugador','jugador','categoria_evento','x','y']]
```

```python
toques = toques.drop(toques[(toques['categoria_evento']=='balón sale')|(toques['categoria_evento']=='decisión contenciosa')|(toques['categoria_evento']=='evento borrado')|
                                  (toques['categoria_evento']=='jugador entra')|(toques['categoria_evento']=='jugador sale')|
                            (toques['categoria_evento']=='mal saque lateral')|(toques['categoria_evento']=='offside provocado')|
                           (toques['categoria_evento']=='portero en penal')|(toques['categoria_evento']=='tarjeta')|
                           (toques['categoria_evento']=='tiro de esquina')|(toques['categoria_evento']=='cubre balón con cuerpo')|
                            (toques['categoria_evento']=='salida despeje de portero')].index)
```

```python
toques.head()
```

```python
toques.categoria_evento.unique()
```

```python
toques[toques['categoria_evento']=='tiro salvado'][['x','y']].describe()
```

```python
toques['zona_x'] = np.where((0<=toques['x'])&(toques['x']<25),'1', 
                                         np.where((25<=toques['x'])&(toques['x']<50),'2',
                                                  np.where((50<=toques['x'])&(toques['x']<75),'3',
                                                           np.where((75<=toques['x'])&(toques['x']<=100),'4',False))))
```

```python
toques['zona_y'] = np.where((0<=toques['y'])&(toques['y']<25),'1', 
                                         np.where((25<=toques['y'])&(toques['y']<50),'2',
                                                  np.where((50<=toques['y'])&(toques['y']<75),'3',
                                                           np.where((75<=toques['y'])&(toques['y']<=100),'4',False))))
```

```python
toques['zona_xy'] = toques['zona_x']+'-'+toques['zona_y']
```

```python
toques.head()
```

```python
toques.zona_xy.unique()
```

```python
#toques['mediocampo'] = np.where((toques['zona_xy']=='2-3')|(toques['zona_xy']=='3-3')|(toques['zona_xy']=='2-2')|
 #                              (toques['zona_xy']=='3-2'),True,False)
```

```python
toques['mediocampo'] = np.where((toques['zona_xy']=='3-1')|(toques['zona_xy']=='3-2')|(toques['zona_xy']=='3-3')|
                                (toques['zona_xy']=='3-3'),True,False)
```

```python
toques.head()
```

```python
toques = toques[['fecha','equipo','id_jugador','jugador','mediocampo']]
```

```python
toques = toques.groupby(['equipo','id_jugador','jugador']).agg(['count','sum']).reset_index().drop('fecha',1)
```

```python
toques.head()
```

```python
toques.columns = toques.columns.get_level_values(0)
```

```python
toques.head()
```

```python
toques.columns.values[3] = 'total'
```

```python
toques.head()
```

```python
toques['ind_mediocampo'] = toques['mediocampo']/toques['total']
```

```python
toques.head()
```

```python
feat11 = toques[['equipo','id_jugador','jugador','ind_mediocampo']]
```
